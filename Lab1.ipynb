```python
import pandas as pd
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plt.scatter(data['Age'],data.Insulin,color='red')
plt.scatter(data['Age'],data.Insulin,color='green')
plt.show()
```

```python
plt.scatter(data['Age'],data.Glucose)
plt.show()
```

```python
plt.scatter(data['Age'],data.Pregnancies)
plt.show()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Glucose'], kde=True, color='skyblue')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=data )
plt.show()
```

```python
data.boxplot(by='Outcome',column=['BMI'])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outcome', y='BMI', data=data,palette="Set1")
plt.title('BMI by Outcome')
plt.xlabel('Outcome')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
